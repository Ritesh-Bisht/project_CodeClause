```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

LOAD THE IMAGE

```python
image=cv2.imread('c:/Users/Ritesh Bisht/Downloads/istockphoto-1218498460-1024x1024.jpg')
```

Display the Image

```python
# Display the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

CONVERT TO GRAY SCALE

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

INVERTING THE GRAYSCALE IMAGE

```python
# Invert the grayscale image
inverted_image = cv2.bitwise_not(gray_image)

# Display the inverted image
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted Image')
plt.axis('off')
plt.show()
```

Apply Gaussian Blur

```python

blurred_image = cv2.GaussianBlur(inverted_image, ( 41, 41), 0)
```

Creating Sketch

```python
# Invert the blurred image
inverted_blurred_image = cv2.bitwise_not(blurred_image)

# Create the sketch by blending the grayscale and inverted blurred images
sketch = cv2.divide(gray_image, inverted_blurred_image, blurred_image, scale=256)

# Display the final sketch
plt.imshow(sketch, cmap='gray')
plt.title('Sketch')
plt.axis('off')
plt.show()
```
